```python
import pandas as pd

orbital_energies = pd.read_csv('t.csv')
#print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Replace these values with the eigenvalues extracted from your sodium.log
#orbital_energies = [-0.986, -0.421, -0.222, 0.076, 0.096, 0.112, 0.212, 0.214, 0.215, 0.217, 0.219, 0.221]

# Create a histogram of the orbital energies
num_bins = 50
dos, bin_edges = np.histogram(orbital_energies, bins=num_bins, density=True)

# Plot the DOS
plt.figure(figsize=(6, 4))
plt.plot(bin_edges[:-1], dos, '-o')
plt.xlabel('Energy (a.u.)')
plt.ylabel('Density of States')
plt.title('Density of States (DOS) for Sodium Atom')
plt.grid(True)
plt.show()
```
